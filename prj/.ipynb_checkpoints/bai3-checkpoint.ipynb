## Câu 3: Phân tích và trực quan hóa dữ liệu về tình hình hoạt động của một nhà mạng di động
Cho tập tin "Telecommunication_data.xlsx" chứa dữ liệu mẫu về các chỉ tiêu KPI (Key Performance Indicator) của một nhà mạng viễn thông. Dữ liệu được ghi nhận theo mức ngày từ năm tháng 1/2022 đến tháng 4/2023.

Trong đó:
- "TRAFFIC2G (Erlang)": lưu lượng thoại (nghe, gọi) trên mạng 2G
- "TRAFFIC3G (Erlang)": lưu lượng thoại trên mạng 3G
- "TRAFFIC VoLTE (Erlang)": lưu lượng thoại trên mạng 4G
- "DATA2G(GB)": lưu lượng dữ liệu (các gói cước data đăng ký để dùng internet trên di động) 2G
- "DATA3G(GB)": lưu lượng dữ liệu (các gói cước data đăng ký để dùng internet trên di động) 3G
- "DATA4G(GB)": lưu lượng dữ liệu (các gói cước data đăng ký để dùng internet trên di động) 4G

Đây là các chỉ tiêu chính để đánh giá mức độ kinh doanh, bán hàng của công ty này.

## Gợi ý:
- Giả sử 6 chỉ tiêu bán hàng chính có đơn giá bán hàng là như nhau. Tổng doanh thu của công ty sẽ là tổng 6 chỉ tiêu KPI chính nhân với đơn giá bán hàng. Và giả sử đơn giá bán hàng không đổi trong các năm
- Theo xu hướng hiện đại hóa, công nghệ hóa, thì tỉ trọng đóng góp vào doanh thu của mạng 2G, 3G nên ngày càng giảm đi và tỉ trọng doanh thu của mạng 4G nên ngày càng tăng lên.

## Giải thích tham số:
- CELLS: số lượng "cell" của mạng 2G                  
- CELLBS: số lượng "cell" của mạng 3G                    
- ECELLS: số lượng "cell" của mạng 4G                    
- SITES: số lượng trạm của mạng 2G, một trạm thì có 3 cell (có 1 ít trạm có ít hơn)                    
- NODEBS: số lượng trạm của mạng 3G, một trạm thì có 3 cell (có 1 ít trạm có ít hơn)                   
- ENODEBS: số lượng trạm của mạng 4G, một trạm thì có 6 cell (có 1 ít trạm có ít hơn)
* Lưu ý: 1 trạm phát sóng nếu có phát sóng 2G thì nhất định có phát sóng 3G và 4G. Tuy nhiên 1 trạm 4G thì không nhất thiết phải có phát sóng 2G.

```python
#Dùng tứ phân vị để kiểm tra giá trị ngoại lệ
import numpy as np
def check_outliers(name):
    we=df[name].values
    Q1=np.quantile(we,0.25)
    Q3=np.quantile(we,0.75)
    IQR=Q3-Q1
    print("Q1: ",Q1,"Q3: ",Q3,"IQR:",IQR)
    outliers=np.unique(we[[(n<Q1-1.5*IQR)|(n>Q3+1.5*IQR) for n in we]])
    indexes=[we.tolist().index(i) for i in outliers]
    print("Indexes of outliers: ", indexes)
    print("Outlier values", we[indexes])
```

```python
for col in numeric_columns:
    check_outliers(col)
```

```python
df['tuần']=df['Ngày'].dt.week
df['tháng']=df['Ngày'].dt.month
df['năm']=df['Ngày'].dt.year
df
```

```python
monthly_data = df.groupby(['năm', 'tháng'])[['DATA2G(GB)', 'DATA3G(GB)','DATA4G(GB)']].sum().reset_index()
monthly_data['Tháng-Năm']=monthly_data['tháng'].astype(str) + '-' + monthly_data['năm'].astype(str)
monthly_data
```

```python
monthly_data.plot(x='Tháng-Năm',y=['DATA2G(GB)', 'DATA3G(GB)','DATA4G(GB)'],use_index=True,title="Biểu đồ trên thể hiện lưu lượng dữ liệu (trong GB) của 2G, 3G và 4G theo tháng ",grid=True,rot=60)
```

Lưu lượng dữ liệu 4G là cao nhất và có xu hướng tăng liên tục.

```python
monthly_data['Total Data (GB)'] = monthly_data[['DATA2G(GB)', 'DATA3G(GB)','DATA4G(GB)']].sum(axis=1)

# Calculate the percentage contribution of each network type to the revenue
monthly_data['2G %'] = (monthly_data['DATA2G(GB)'] / monthly_data['Total Data (GB)']) * 100
monthly_data['3G %'] = (monthly_data['DATA3G(GB)'] / monthly_data['Total Data (GB)']) * 100
monthly_data['4G %'] = (monthly_data['DATA4G(GB)'] / monthly_data['Total Data (GB)']) * 100
monthly_data.plot(x="Tháng-Năm", y=["2G %", "3G %", "4G %"], kind='bar', stacked=True, 
                  title="Tỉ lệ % đóng góp vào doanh thu của lưu liệu mạng của mạng 2G, 3G và 4G", grid=True, rot=60, figsize=(15, 7))
```

Nhận xét, tỉ lệ đóng góp 2G hầu như không đáng kể, số liệu cho thấy 4G, chiếm vị thế quan trọng trong việc đóng góp vào doanh thu của công ty

```python
month_revenue=df.groupby(['năm','tháng'])[['DATA2G(GB)', 'DATA3G(GB)','DATA4G(GB)','TRAFFIC2G(Erlang)','TRAFFIC3G(Erlang)','TRAFFIC VoLTE(Erlang)']].sum().reset_index()
month_revenue['Tháng-Năm']=month_revenue['tháng'].astype(str) + '-' + month_revenue['năm'].astype(str)
month_revenue['Total']=month_revenue[['DATA2G(GB)', 'DATA3G(GB)','DATA4G(GB)','TRAFFIC2G(Erlang)','TRAFFIC3G(Erlang)','TRAFFIC VoLTE(Erlang)']].sum(axis=1)
month_revenue.plot(x='Tháng-Năm',y='Total',grid=True,rot=60,legend=True,title="Biểu đồ doanh thu trong năm 2022,2023")

```

```python
month_rev_each=month_revenue
month_rev_each['2G %'] = ((month_rev_each['DATA2G(GB)']+month_rev_each['TRAFFIC2G(Erlang)']) / month_rev_each['Total']) * 100
month_rev_each['3G %'] = ((monthly_data['DATA3G(GB)']+month_rev_each['TRAFFIC3G(Erlang)']) / month_rev_each['Total']) * 100
month_rev_each['4G %'] = ((monthly_data['DATA4G(GB)']+month_rev_each['TRAFFIC VoLTE(Erlang)']) / month_rev_each['Total']) * 100
month_rev_each.plot(x="Tháng-Năm", y=["2G %", "3G %", "4G %"], kind='bar', stacked=True, 
                  title="Tỉ lệ % đóng góp vào doanh thu của mạng 2G, 3G và 4G năm 2022,2023", grid=True, rot=60, figsize=(15, 7))
```

```python
Biểu đồ trên cho thấy được tỉ lệ % doanh thu đóng góp vào tổng doanh thu năm 2022, 2023.Nhìn vào biểu đồ trên ta thấy được tỉ lệ doanh thu 4G chiếm phần lớn trong tổng doanh thu của công ty năm 2022,2023 và có xu hướng tăng mạnh về sau, tỉ lệ 3G chiếm phần tương đối ổn định và tỉ lệ 2G chỉ chiếm phần nhỏ trong cơ cấu tổng doanh thu
```
